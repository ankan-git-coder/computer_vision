```python
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread("messi.jpg")

cv2_imshow(img)
```

```python
cv2.resize(img,(640,480))
```

```python
cv2_imshow(img)
```

```python
cropped_img = img[100:300,150:300]
cv2_imshow(cropped_img)
```

```python
IMG_COLOURSPACE = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(IMG_COLOURSPACE)
```

```python
IMG_COLOURSPACE = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2_imshow(IMG_COLOURSPACE)
```

```python
IMG_COLOURSPACE = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
cv2_imshow(IMG_COLOURSPACE)
```

```python
# BLURRING AN IMAGE
```

```python
img_blur = cv2.blur(img,(5,5))
cv2_imshow(img_blur)
```

```python
median = cv2.medianBlur(img, 5)

cv2_imshow(median)
```

```python
gaussian = cv2.GaussianBlur(img, (5,5), 0)

cv2_imshow(gaussian)
```

```python
ret, thresh = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
cv2_imshow(thresh)
```
